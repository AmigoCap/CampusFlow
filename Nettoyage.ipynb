```python
import pandas as pd
import glob
import gmaps
import gmaps.datasets
from datetime import datetime
from datetime import timedelta
from dateutil.parser import parse
```

```python
path = "Data/Privamov/"
allFiles = glob.glob(path + "/*")
frame = pd.DataFrame()
list_= []

for file_ in allFiles:
    df = pd.read_csv(file_,sep="\t", header=None,  names = ["Id", "DateTime", "Longitude", "Latitude"])
    list_.append(df)
frame = pd.concat(list_)
```

```python
ref_time = frame.iloc[0]['DateTime']
#print(current_time)
for x in frame.itertuples():
    current_time = x.DateTime
    if (ref_time > current_time):
        print("Plus grand !")
    ref_time = x.DateTime
```
